```python
import csv
with open('mpg.csv') as csvfile:
    mpg= list(csv.DictReader(csvfile))
print(mpg[:3])
print(len(mpg))
```

```python
mpg[0].keys()
```

```python
mpg[0].values()
```

```python
sum(float(d['cty']) for d in mpg)/ len(mpg)
```

```python
City_mpg_by_cyl=[]
cylinders= set(d['cyl'] for d in mpg)
for c in cylinders:
    sumcty=0
    cylcount=0
    for d in mpg:
        if d['cyl']== c:
            sumcty += float(d['cty'])
            cylcount += 1
    City_mpg_by_cyl.append((c, sumcty/cylcount))
City_mpg_by_cyl.sort(key= lambda x:x[0])
City_mpg_by_cyl
```

```python
people = ['Dr. Christopher Brooks', 'Dr. Kevyn Collins-Thompson', 'Dr. VG Vinod Vydiswaran', 'Dr. Daniel Romero']

def split_title_and_name(people1):
    return people1.split()[0] + ' '+ people1.split()[-1]
list(map(split_title_and_name , people))
```

```python
def split_title_and_name(person):
    title= person.split()[0]
    lname= person.split()[-1]
    return '{} {}'.format(title,lname)

list(map(split_title_and_name, people))
```

### Distributions in Pandas

```python
import pandas as pd
import numpy as np

chance_of_tornado=.01
tornado_events= np.random.binomial(1,chance_of_tornado, 1000000)

tornado_in_two_days_a_row=0
for i in range(1, len(tornado_events)-1):
    if tornado_events[i]==1 and tornado_events[i-1]:
        tornado_in_two_days_a_row +=1
print(tornado_in_two_days_a_row)
```
