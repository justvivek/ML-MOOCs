# GoodHart's Law

### Goodhart's law states that "When a measure becomes a target, it ceases to be a good measure". In this we'll discuss how this affects SEO, call centers, and Scrum.

### Cobras effect/ Rat effect law:- The cobra effect occurs when an attempted solution to a problem actually makes the problem worse.

### In vietnam-- to get rid of rat--gov crowdsource the work--bring the tail of the rat and get money-- observation:- seeing a lot of tailless rats and people start breeding rat to chop off their tails and get money

### In India-- gov pay to kill the snakes and people start breeding snakes, so gov stopped it and people released all the snakes


### Campbell's Law is as follows: The more any quantitative social indicator is used for social decision-making, the more subject it will be to corruption pressures and the more apt it will be to distort and corrupt the social processes it is intended to monitor.(Allocating fund to clean the municipal areas--> for this authority do inspection --> local leaders do more littering to get more funds)


### When heavily focussed on optimizing one metric or KPI(key performance indicator--defined at the start of the project for example no. of subscription, profit, no. of producs sold---> it is important to measure but if you only measure that then that is not good), it ceases to become right KPI, people start manipulating it.
### if you improve a product more than enough eventually you will break it
### Call centre: try to reduce the time on call, so come up with better script, menu navigation etc. but it'll also encourage the employees to quick hangup the phone, not help the customer properly, drop calls might also increase

### SEO:- GoodHarts law is there in place thatswhy they have to keep it secret now how they rank the pages on google, (link farms, hidden data dictionary in web pages)

### GoodHarts law is always look out for, so lot of data science is picking up a metric or KPI and optimizing for it, as soon as we picked for it and start optimizing it, there is a chance that you might be affecting or changing the way people are think/behave, which could then change the usefullness of the change we are measuring

### Agile/Scrum Model in companies:-  for some companies it is good, for others it might not be productive at all

```python
staff_df = pd.DataFrame([{'Name': 'Kelly', 'Role': 'Director of HR'},
                         {'Name': 'Sally', 'Role': 'Course liasion'},
                         {'Name': 'James', 'Role': 'Grader'}])
staff_df= staff_df.set_index('Name')
student_df= pd.DataFrame([{'Name': 'James', 'School': 'Business'},
                           {'Name': 'Mike', 'School': 'Law'},
                           {'Name': 'Sally', 'School': 'Engineering'}]) 
student_df= student_df.set_index('Name')
print(staff_df)
print()
print(student_df)
```

```python
staff_df = pd.DataFrame([{'First Name': 'Kelly', 'Last Name': 'Desjardins', 'Role': 'Director of HR'},
                         {'First Name': 'Sally', 'Last Name': 'Brooks', 'Role': 'Course liasion'},
                         {'First Name': 'James', 'Last Name': 'Wilde', 'Role': 'Grader'}])
student_df = pd.DataFrame([{'First Name': 'James', 'Last Name': 'Hammond', 'School': 'Business'},
                           {'First Name': 'Mike', 'Last Name': 'Smith', 'School': 'Law'},
                           {'First Name': 'Sally', 'Last Name': 'Brooks', 'School': 'Engineering'}])
staff_df
student_df
#how='outer' in this we take union of dataframes based on the selected index
pd.merge(staff_df, student_df,how='outer',left_on=['First Name','Last Name'], right_on=['First Name','Last Name'])
```

```python
#how='inner' in this we take intersection of dataframes based on the selected index
pd.merge(staff_df, student_df,how='inner',left_on=['First Name','Last Name'], right_on=['First Name','Last Name'])
```

```python
#how='left' left outer join
pd.merge(staff_df, student_df,how='left',left_on=['First Name','Last Name'], right_on=['First Name','Last Name'])
```

```python
##how='right' right outer join
pd.merge(staff_df, student_df,how='right',left_on=['First Name','Last Name'], right_on=['First Name','Last Name'])
```

```python
df= df[df['SUMLEV']==50]
df.set_index(['STNAME','CTYNAME'],inplace=True)
df.rename(columns={'ESTIMATES2010':'ESTIMATES BASE 2010'})
```

```python
import numpy as np
def min_max(row):
    data= row[['POPESTIMATE2010',
                'POPESTIMATE2011',
                'POPESTIMATE2012',
                'POPESTIMATE2013',
                'POPESTIMATE2014',
                'POPESTIMATE2015']]
    return pd.Series({'max':np.max(data),'min':np.min(data)})
df.apply(min_max,axis=1)
```

```python
import numpy as np
def min_max(row):
    data= row[['POPESTIMATE2010',
                'POPESTIMATE2011',
                'POPESTIMATE2012',
                'POPESTIMATE2013',
                'POPESTIMATE2014',
                'POPESTIMATE2015']]
    row['max']=np.max(data)
    row['min']=np.min(data)
    return row
#     return pd.Series({'max':np.max(data),'min':np.min(data)})
df.apply(min_max,axis=1)
```

# Group By

```python
import pandas as pd
import numpy as np
df= pd.read_csv('census.csv')
df= df[df['SUMLEV']==50]
df
```

```python
%%timeit -n 10
for state in df['STNAME'].unique():
    avg= np.average(df.where(df['STNAME']==state).dropna()['CENSUS2010POP'])
    print('Counties in State '+ state +' have an average population of '+str(avg))
```

```python
# df_g=df.groupby('STNAME')
df_g= df.set_index(['STNAME','CTYNAME'])
# df_g= df_g.groupby(df_g.index)
df_g
```

```python
%%timeit -n 10
for group, frame in df.groupby('STNAME'):
    avg= np.average(frame['CENSUS2010POP'])
    print('Counties in State '+ group +' have an average population of '+str(avg))
```

```python
df= pd.read_csv('census.csv')
df= df[df['SUMLEV']==50]
df= df.set_index('STNAME')
def fun(item):
    if item[0]<'M':
        return 0
    if item[0]<'Q':
        return 1
    return 2
for group, frame in df.groupby(fun):
    print('There are ' + str(len(frame)) + ' records in group ' + str(group) + ' for processing')
```
