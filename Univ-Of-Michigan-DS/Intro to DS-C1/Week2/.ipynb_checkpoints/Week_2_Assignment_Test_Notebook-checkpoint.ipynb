```python
import pandas as pd

df= pd.read_csv('olympics.csv', index_col=0, skiprows=1)
df.head()
for col in df.columns:
    if col[:2]=='01':
        df.rename(columns={col:'Gold' + col[4:]}, inplace=True)
    if col[:2]=='02':
        df.rename(columns={col:'Silver' + col[4:]}, inplace=True)
    if col[:2]=='03':
        df.rename(columns={col:'Bronze' + col[4:]}, inplace=True)
    if col[:1]=='â„–':
        df.rename(columns={col:'#' + col[1:]}, inplace=True)
# df.head()
name_ids= df.index.str.split('\s\(')
df.index= name_ids.str[0]
df['ID']= name_ids.str[1].str[0:3]
df=df.drop('Totals')

df.head()


```

```python
# Countries who has won atleast one Gold in summer and winter games
atleast1gold = df[(df['Gold']>=1) & (df['Gold.1']>=1)]
atleast1gold
```

```python
relative_max_gold= ((atleast1gold['Gold']-atleast1gold['Gold.1'])/atleast1gold['Gold.2'])
relative_max_gold
```

```python
atleast1gold[relative_max_gold == relative_max_gold.max()].index[0]
```

```python
df['Points']= df['Gold.2']*3 + df['Silver.2']*2 + df['Bronze.2']
df['Points']
```
