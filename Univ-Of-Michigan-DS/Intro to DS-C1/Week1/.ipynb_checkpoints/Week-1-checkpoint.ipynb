## Reading and Writing CSV files

Let's import our datafile mpg.csv, which contains fuel economy data for 234 cars.

mpg : miles per gallon
class : car classification
cty : city mpg
cyl : # of cylinders
displ : engine displacement in liters
drv : f = front-wheel drive, r = rear wheel drive, 4 = 4wd
fl : fuel (e = ethanol E85, d = diesel, r = regular, p = premium, c = CNG)
hwy : highway mpg
manufacturer : automobile manufacturer
model : model of car
trans : type of transmission
year : model year

```python
import csv
%precision 2
with open('mpg.csv') as csvfile:
    mpg= list(csv.DictReader(csvfile))
mpg[:3] # The first three dictionaries in our list.
```

```python
# csv.Dictreader has read in each row of our csv file as a dictionary. len shows that our list is comprised of 234 dictionaries.
len(mpg)
```

```python
mpg[0].keys()
```

```python
mpg[0].values()
```

```python
# to find the average cty fuel economy across all cars. All values in the dictionaries are strings, so we need to convert to float.
sum(float(d['cty']) for d in mpg)/ len(mpg)
```

```python
# Similarly this is how to find the average hwy fuel economy across all cars.
sum(float(d['hwy']) for d in mpg)/ len(mpg)
```

```python
# Use set to return the unique values for the number of cylinders the cars in our dataset have.
cylinders= set(d['cyl'] for d in mpg)
print(cylinders)
```

### grouping the cars by number of cylinder, and finding the average cty mpg for each group.

```python
City_mpg_byCyl=[]

for c in cylinders:
    sumpg=0
    cyltypecount=0
    for d in mpg:
        if d['cyl']== c:
            sumpg += float(d['cty'])
            cyltypecount +=1
    City_mpg_byCyl.append((c,sumpg / cyltypecount))
City_mpg_byCyl.sort(key= lambda x:x[0])
City_mpg_byCyl
```

```python
# Use set to return the unique values for the class types in our dataset.
vehicleclass= set(d['class'] for d in mpg)
vehicleclass
```

### And here's an example of how to find the average hwy mpg for each class of vehicle in our dataset.

```python
hwy_mpg_byClass= []

for c in vehicleclass:
    sumpg= 0
    vclasscount=0
    for d in mpg:
        if d['class']== c:
            sumpg += float(d['hwy'])
            vclasscount += 1
    hwy_mpg_byClass.append((c, sumpg/vclasscount))
hwy_mpg_byClass.sort(key=lambda x: x[1])
hwy_mpg_byClass
```

### Writing function as lambdas

```python

people = ['Dr. Christopher Brooks', 'Dr. Kevyn Collins-Thompson', 'Dr. VG Vinod Vydiswaran', 'Dr. Daniel Romero']

def split_title_and_name(person):
    return person.split()[0] + ' ' + person.split()[-1]

#option 1
for person in people:
    print(split_title_and_name(person) == (lambda x: x.split()[0] + ' ' + x.split()[-1])(person))

#option 2
#list(map(split_title_and_name, people)) == list(map(lambda person: person.split()[0] + ' ' + person.split()[-1], people))
```

```python
# iterate from 0 to 41 and return the even numbers.
my_list=[]
for i in range(0,41):
    if (i%2)==0:
        my_list.append(i)
my_list
```

```python
# Now the same thing but with list comprehension.
my_list=[i for i in range(1,41) if i%2 ==0]
my_list
```

```python
# Converting function into a list comprehension
def times_tables():
    lst = []
    for i in range(10):
        for j in range (10):
            lst.append(i*j)
    return lst

times_tables() == [i*j for j in range(10) for i in range(10)]

# lst= [i*j for i in range(10) for j in range(10)]
# lst
```

## Indexing / Slicing

```python
s= np.arange(13)**2
s
```

```python
# Use bracket notation to get the value at a specific index. Remember that indexing starts at 0.
print(s[0],s[4],s[-1])
# Use : to indicate a range. array[start:stop]
# Leaving start or stop empty will default to the beginning/end of the array
# Use negatives to count from the back.
print(s[1:5])
print(s[:5])
print(s[5:])
print(s[-4:])
print(s[:-5])
```

```python
# A second : can be used to indicate step-size. array[start:stop:stepsize]
# Here we are starting 5th element from the end, and counting backwards by 2 until the beginning of the array is reached.
print(s[-5::-2])
```

## Copying Data
### Be careful with copying and modifying arrays in NumPy!
### r2 is a slice of r

```python
r2= r[:3,:3]
r2
```

```python
# Set this slice's values to zero ([:] selects the entire array)
r2[:]=0
r2
```

```python
# r has also been changed!
r
```

```python
# To avoid this, use r.copy to create a copy that will not affect the original array
r= np.arange(36)
r.resize(6,6)
r_copy= r.copy()
r3= r_copy[:3,:3]
r3
```

```python
r3[:]=0
r3
```

```python
# Now when r_copy is modified, r will not be changed.
print(r_copy,'\n')
print(r)
```
